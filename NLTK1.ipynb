```python
import nltk # import the nltk module
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize # import word_tokenize from nltk.tokenize

# Download 'punkt_tab' data package if not already downloaded
nltk.download('punkt_tab')

sentence = "It is good to be a pythoner who is pythonly pythoning with python"
words = word_tokenize(sentence) # use word_tokenize directly
for word in words:
    ps = PorterStemmer()
    print(ps.stem(word))
```
